# EastWestAirlines

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
xls = pd.ExcelFile('EastWestAirlines.xlsx')
air = pd.read_excel(xls,'data')
air
```

```python
air.info()
```

```python
air1=air.drop(['ID#'],axis=1)
air1
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame 
air1_norm = norm_func(air1.iloc[:,1:])
```

```python
air1_norm.head()
```

# Hierarchical

```python
# create dendrogram
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(air1_norm, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
hc
```

```python
# save clusters for chart
y_hc = hc.fit_predict(air1_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
#Creating a new column with clusterid and including it in Al_norm dataset
air1_norm['Clusterid']= hc.labels_
air1_norm
```

# kmeans

```python
air1_norm
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change. 
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(air1_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(air1_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
air4=air1.copy()
air4['clusters4id']=clusters4.labels_
air4
```

```python
# Compute the centroids for K=4 clusters with 11 variables
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
air4.groupby('clusters4id').agg(['mean']).reset_index()
```
